```python
from pymatgen.ext.matproj import MPRester
from pymatgen.analysis.diffraction.xrd import XRDCalculator
```

```python
with MPRester("0xKCQrLrxyfoPKG3N") as m:
    structure = m.get_structure_by_material_id("mp-7477")
xdrcalculator = XRDCalculator(1.54060, symprec=1) #argument symprec=1 pour que xdrcalculator donne les indices de Bravais-Miller car structure R
pattern = xdrcalculator.get_pattern(structure)
```

```python
pics = ['Premier', 'Deuxième', 'Troisième']
for i in range(3) : 
    print("{} pic : \n Intensité : {} (scaled) \n Angle : {} (2 theta) \n Indices : {} (de multiplicité {})".format(pics[i], pattern.y[i], pattern.x[i], pattern.hkls[i][0]['hkl'], pattern.hkls[i][0]['multiplicity']))
    print("---------------")
```

```python
xdrcalculator.show_plot(structure)
```
